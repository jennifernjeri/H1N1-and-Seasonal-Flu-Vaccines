**`Handling employment_industry and employment_occupation missing data`**

Instead of dropping the missing values, which would significantly impact about half of the dataset crucial for modeling, we have chosen to fill in the missing values using the available distribution in each column. This approach allows us to retain valuable data for analysis and modeling while addressing the missing values. As we fill nas we will keep in mind that half of the values in these columns are generated.

The values in these columns are coded this shows that the data engineering team adhered to Data Privacy.


```python
# Check current status of missing values
vaccine_df.isna().mean()
```

```python
# Check distribution of value_counts
distribution = vaccine_df['employment_industry'].value_counts()
distribution
```

```python
# Normalize the distribution
total_values = sum(distribution.values)
normalized_distribution = {key: value / total_values for key, value in distribution.items()}

# Fill missing values with the same distribution
vaccine_df['employment_industry'] = vaccine_df['employment_industry'].fillna(
    pd.Series(np.random.choice(list(normalized_distribution.keys()),
                                p=list(normalized_distribution.values()),
                                size=len(vaccine_df)))
)

print('Missing values after distribution fillna:', vaccine_df['employment_industry'].isna().mean())

# Fill remaining missing values with the mode
vaccine_df['employment_industry'] = vaccine_df['employment_industry'].fillna(vaccine_df['employment_industry'].mode().iloc[0])

print('Missing values after Mode fillna:', vaccine_df['employment_industry'].isna().mean())
```

```python
# View new distribution
print(vaccine_df['employment_industry'].value_counts())
```

**`Handling missing values in employment_occupation column`**

Utilizing domain knowledge, it's observed that employment occupation is closely aligned with employment industry. Consequently, we plan to correlate combinations of industry and occupations. This strategy will be instrumental in filling missing values more precisely, harnessing the inherent connection between these two variables.

```python
# Create a mapping dictionary based on available combinations between industry and occupation
mapping_dict_occupation = vaccine_df.groupby('employment_industry')['employment_occupation'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan).to_dict()

# Apply the mapping to fill missing values in the employment occupation column
vaccine_df['employment_occupation'] = vaccine_df['employment_occupation'].fillna(vaccine_df['employment_industry'].map(mapping_dict_occupation))

print('Missing values after combination fillna:', vaccine_df['employment_occupation'].isna().mean())
```

###**Evaluate the models**

```python
class ModelEvaluator:
    def __init__(self, pipeline, X_test, y_test):
        self.pipeline = pipeline
        self.X_test = X_test
        self.y_test = y_test
        self.fpr = None
        self.tpr = None
        self.auc = None

    def evaluate_model(self):
        y_pred = self.pipeline.predict(self.X_test)
        y_prob = self.pipeline.predict_proba(self.X_test)[:, 1]

        # Calculate ROC curve and AUC
        fpr, tpr, _ = roc_curve(self.y_test, y_prob)
        auc_score = auc(fpr, tpr)

        # Store values
        self.fpr = fpr
        self.tpr = tpr
        self.auc = auc_score

        model_name = self.pipeline.named_steps['classifier'].__class__.__name__
        print(f"{model_name} - AUC-ROC: {auc_score:.4f}")

    def plot_roc_curve(self, ax=None):
        if ax is None:
            ax = plt.gca()

        fpr, tpr, _ = roc_curve(self.y_test, self.pipeline.predict_proba(self.X_test)[:, 1])
        auc_score = auc(fpr, tpr)

        # Store values
        self.fpr = fpr
        self.tpr = tpr
        self.auc = auc_score

        model_name = self.pipeline.named_steps['classifier'].__class__.__name__
        ax.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.4f})')

evaluators = []

# Evaluate and store values for all models
for pipeline in pipelines:
    evaluator = ModelEvaluator(pipeline, X_test, y_test)
    evaluator.evaluate_model()
    evaluators.append(evaluator)
```
